```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.preprocessing import image
import numpy as np
import os

# Define a simple CNN
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten()
])

# Function to extract features
def extract_features_cnn(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0  # normalize
    features = cnn_model.predict(img_array)
    return features
```

```python
import os

image_folder = '/content/drive/MyDrive/img1'
features_cnn = []
features_resnet = []
image_names = []

for img_name in os.listdir(image_folder):
    if img_name.lower().endswith(('.jpg', '.jpeg', '.png')):
        path = os.path.join(image_folder, img_name)
        f_cnn = extract_features_cnn(path)
        #f_resnet = extract_features_resnet(path)
        features_cnn.append(f_cnn.flatten())
        #features_resnet.append(f_resnet.flatten())
        image_names.append(img_name)
```

```python
import os
import numpy as np
from tqdm import tqdm

# Your CNN feature extraction function must be defined earlier:
# def extract_features_cnn(image_path): ...

# ✅ Set your image folder path
image_folder = '/content/drive/MyDrive/img1'

# ✅ Make sure old files are removed (optional safety)
for file in ['cnn_features_final.npz', 'cnn_features_checkpoint.npz']:
    if os.path.exists(file):
        os.remove(file)
        print(f"🗑️ Removed old file: {file}")

# ✅ Initialize empty lists to store results
features_cnn = []
image_names = []

# ✅ Get a sorted list of all image files
all_images = sorted([f for f in os.listdir(image_folder)
                     if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

print(f"\n📸 Total images to process: {len(all_images)}\n")

# ✅ Start processing each image
for idx, img_name in enumerate(tqdm(all_images, desc="Processing images")):
    path = os.path.join(image_folder, img_name)

    try:
        # Extract CNN features
        f_cnn = extract_features_cnn(path)

        # Store extracted data
        features_cnn.append(f_cnn.flatten())
        image_names.append(img_name)

    except Exception as e:
        print(f"❌ Error processing {img_name}: {e}")

# ✅ Save final results
np.savez_compressed('cnn_features_final.npz',
                    cnn=np.array(features_cnn),
                    names=np.array(image_names))

print("\n✅ All CNN features extracted and saved to 'cnn_features_final.npz'\n")

# ✅ Print features for each image with dotted line separators
np.set_printoptions(precision=4, suppress=True)

for i in range(len(image_names)):
    print(f"Image {i+1}/{len(image_names)}: {image_names[i]}")
    print("\n--- CNN Features ---")
    print(features_cnn[i])
    print("\n" + "-" * 80 + "\n")
```
