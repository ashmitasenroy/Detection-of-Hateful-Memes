```python
import pytesseract
from PIL import Image
import os

# Function to extract text from an image file
def extract_text_from_image(image_path):
    img = Image.open(image_path)  # Open the image
    text = pytesseract.image_to_string(img)  # Extract text using pytesseract
    return text.strip()  # Strip leading/trailing whitespace

# Directory containing the images (update this path to your image folder)
image_folder = "/content/drive/MyDrive/Dataset/img1"

# Loop through all the files in the directory
for image_file in os.listdir(image_folder):
    # Check if the file has a valid image extension
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # Full path to the image
        image_path = os.path.join(image_folder, image_file)

        # Extract text from the image
        text = extract_text_from_image(image_path)

        # Print the formatted output
        print(f"{'='*40}\n"
              f"Image File: {image_file}\n"
              f"Extracted Text:\n{text if text else 'No text found.'}\n"
              f"{'='*40}")
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import os

# Function to preprocess the image for CNN
def preprocess_image(image_path, target_size=(224, 224)):
    img = load_img(image_path, target_size=target_size)  # Load and resize the image
    img_array = img_to_array(img)  # Convert to NumPy array
    img_array /= 255.0  # Normalize pixel values to the range [0, 1]
    return img_array  # Return the normalized image array

# Directory containing the images (update this path to your image folder)
image_folder = "/content/drive/MyDrive/Dataset/img1"

# Initialize a dictionary to store preprocessed images
preprocessed_images = {}

# Loop through all the files in the directory
for image_file in os.listdir(image_folder):
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        image_path = os.path.join(image_folder, image_file)

        # Preprocess the image for CNN
        processed_image = preprocess_image(image_path)
        preprocessed_images[image_file] = processed_image

# Optionally, print a confirmation message
print("Images preprocessed and stored successfully.")
```

```python
import pytesseract
from PIL import Image
import os
import pandas as pd

# Function to extract text from an image file
def extract_text_from_image(image_path):
    img = Image.open(image_path)  # Open the image
    text = pytesseract.image_to_string(img)  # Extract text using pytesseract
    return text.strip()  # Strip leading/trailing whitespace

# Directory containing the images (update this path to your image folder)
image_folder = "/content/drive/MyDrive/Dataset/img1"

# Initialize a list to store data for the DataFrame
data = []

# Loop through all the files in the directory
for image_file in os.listdir(image_folder):
    # Check if the file has a valid image extension
    if image_file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
        # Full path to the image
        image_path = os.path.join(image_folder, image_file)

        # Extract text from the image
        extracted_text = extract_text_from_image(image_path)

        # Add the image name, extracted text, and a placeholder for label to the data list
        data.append({
            'image_name': image_file,
            'extracted_text': extracted_text if extracted_text else 'No text found.',
            'label': None  # Placeholder for labels
        })

# Create a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('/content/drive/MyDrive/labeled_meme_dataset.csv', index=False)

# Optional: Print a summary of the DataFrame
print(df.head())
```

```python
import pandas as pd
from transformers import pipeline

# Load the CSV file that contains the image names and extracted text (assuming labels are not present)
df = pd.read_csv('/content/drive/MyDrive/labeled_meme_dataset.csv')

# Initialize a sentiment-analysis pipeline (you can use a hate-speech detection model instead)
classifier = pipeline('sentiment-analysis')

# Function to automatically label images using sentiment analysis
def auto_label(row):
    # Analyze the sentiment of the extracted text
    result = classifier(row['extracted_text'])[0]

    # Label as hateful (1) if negative sentiment, non-hateful (0) if positive/neutral
    if result['label'] == 'NEGATIVE':
        return 1  # Hateful
    else:
        return 0  # Non-hateful

# Apply the auto-labeling function to each row in the DataFrame
df['label'] = df.apply(auto_label, axis=1)

# Save the updated DataFrame back to a CSV file
df.to_csv('/content/drive/MyDrive/Full_labeled_meme_dataset.csv', index=False)

# Optional: Print out a summary of the labeling
print(df[['image_name', 'label']].head())
```
